# NumPy Assignment
## Working with 1D Arrays and Statistical Operations

```python
# Import NumPy library
import numpy as np
```

## Step 1: Create a 1D NumPy array with integers from 1 to 20

```python
# Create array with integers from 1 to 20
array = np.arange(1, 21)
print("Array:", array)
```

## Part (a): Calculate sum, mean, median, and standard deviation

```python
# Calculate statistical measures
array_sum = np.sum(array)
array_mean = np.mean(array)
array_median = np.median(array)
array_std = np.std(array)

# Display results
print(f"Sum: {array_sum}")
print(f"Mean: {array_mean}")
print(f"Median: {array_median}")
print(f"Standard Deviation: {array_std}")
```

## Part (b): Find indices of elements greater than 10

```python
# Find indices where elements are greater than 10
indices = np.where(array > 10)[0]

print(f"Indices of elements greater than 10: {indices}")
print(f"Elements at those indices: {array[indices]}")
```

## Summary: All operations in one cell

```python
# Complete solution
import numpy as np

# Create array
array = np.arange(1, 21)
print("Array:", array)
print("\n--- Statistical Measures ---")

# Part (a): Statistical operations
print(f"Sum: {np.sum(array)}")
print(f"Mean: {np.mean(array)}")
print(f"Median: {np.median(array)}")
print(f"Standard Deviation: {np.std(array)}")

# Part (b): Find indices
print("\n--- Elements Greater Than 10 ---")
indices = np.where(array > 10)[0]
print(f"Indices: {indices}")
print(f"Values: {array[indices]}")
```
